```python
body_tablerow = soup.find('tbody').find_all('tr', class_ = 'rrow  age-SEN  ')
```

```python
df = pd.DataFrame(columns=['Rank','Name','Country','Points'])
```

```python
df.columns
```

```python
Rank = []
Rank_up_down = []
Name = []
Country = []
Points = []
count = 0
```

```python
for i in body_tablerow:
    table_data = i.find_all('td')
    count = count + 1
    Rank.append(count)
    Name.append(table_data[1].text)
    Country.append(table_data[2].text)
    Points.append(table_data[3].text)
```

```python
frame = {'Rank':Rank,'Name':Name,'Country':Country,'Points':Points}
```

```python
df = pd.DataFrame(frame)
```

```python
df.to_excel('Table_tennis_data.xlsx', encoding='utf-8')
```
