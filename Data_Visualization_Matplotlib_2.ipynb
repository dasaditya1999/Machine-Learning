### Draw bar Plot using bar()

```python
fig = plt.figure()
ax = fig.add_subplot(111)

covid_year = ['2019','2020','2021','2022']
male_affected = [1,1.5,1.75,0.5]
female_affected = [1.6,2,2.3,0.6]

index = np.arange(len(covid_year))

ax.bar(index - 0.2, male_affected,color='b', label='male')
ax.bar(index + 0.2, female_affected, color='g', label='female')

plt.xticks(index,covid_year)

ax.set(xlabel='Covid Year', ylabel='No. of imfected People in CR', title='Covid Analysis')

ax.legend()
```

### Subplots Using 'GridSpec'

+ GridSpec class of matplotlib.gridspec can also be used to create Subplots.

+ Initially, a grid with given number of rows and columns is set up.

+ Later while creating a subplot, the number of rows and columns of grid, spanned by the subplot are provided as inputs to subplot function.

```python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,8))
gd = gridspec.GridSpec(2,2)
axes1 = plt.subplot(gd[0,:],title='Plot1')
axes1.set_xticks([]); axes1.set_yticks([])
axes2 = plt.subplot(gd[1,0],title = 'plot2')
axes2.set_xticks([]); axes2.set_yticks([])
axes3 = plt.subplot(gd[1,-1],title='plot3')
axes3.set_xticks([]); axes3.set_yticks([])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

def generate_figure1():

    # Write your functionality below
    x = np.arange(0.0,10.0,0.01)
    arr_s1 = np.sin(3*np.pi*x)
    arr_s2 = np.sin(6*np.pi*x)
    
    fig = plt.figure(figsize=(12,7))
    axes1 = plt.subplot(2,1,1)
    axes1.plot(x,arr_s1)
    axes1.set(title='Sin(3*pi*x)')
    
    axes2 = plt.subplot(2,1,2)
    axes2.plot(x,arr_s2)
    axes2.set(title='Sin(6*pi*x)')

    return fig


def generate_figure2():

    # Write your functionality below
    np.random.seed(1500)
    x = np.random.rand(10)
    y = np.random.rand(10)
    z = np.sqrt(x**2+y**2)
    
    fig = plt.figure(figsize=(9,7))
    
    fig = plt.figure(figsize=(9,7))

    axes1 = plt.subplot(2,2,1)
    axes1.scatter(x=x,y=y,s=80,c=z, marker='d')
    axes1.set(title='Scatter plot with Diamond Markers',xticks=[0.0,0.5,1.0,1.5], yticks=[-0.2,0.2,0.6,1.0])

    axes2 = plt.subplot(2,2,2)
    axes2.scatter(x,y,s=80,c=z, marker='o')
    axes2.set(title='Scatter plot with Circle Markers',xticks=[0.0,0.5,1.0,1.5], yticks=[-0.2,0.2,0.6,1.0])

    axes3 = plt.subplot(2,2,3)
    axes3.scatter(x,y,s=80,c=z, marker='+')
    axes3.set(title='Scatter plot with Plus Markers',xticks=[0.0,0.5,1.0,1.5], yticks=[-0.2,0.2,0.6,1.0])

    axes4 = plt.subplot(2,2,4)
    axes4.scatter(x,y,s=80,c=z, marker='^')
    axes4.set(title='Scatter plot with Upper Triangle Markers',xticks=[0.0,0.5,1.0,1.5], yticks=[-0.2,0.2,0.6,1.0])

    plt.tight_layout()

    return fig


def generate_figure3():

    # Write your functionality below
    x = np.arange(1,301,3)
    y1 = x
    y2 = x**2
    y3 = x**3
    
    fig = plt.figure(figsize=(9,7))
    gr = gridspec.GridSpec(2,2)

    #First row & first Column
    axes1 = plt.subplot(gr[0,0],title='y = x')
    axes1.plot(x,y1)

    #Second row & first column
    axes2 = plt.subplot(gr[1,0],title='y = x**2')
    axes2.plot(x,y2)

    #Second column both the rows
    axes3 = plt.subplot(gr[:,1],title='y = x**3')
    axes3.plot(x,y3)
    
    plt.tight_layout()

    return fig


# Note: Do Not modify the below code
if __name__ == '__main__':
    import sys
    import subprocess
    subprocess.check_call(
        [sys.executable, "-m", "pip", "install", 'test_plot6-0.2-py3-none-any.whl'], stdout=subprocess.DEVNULL)
    from test_plot6 import matplotlib6
    usr_fig1 = generate_figure1()
    usr_fig2 = generate_figure2()
    usr_fig3 = generate_figure3()
    matplotlib6.save_answer(usr_fig1, usr_fig2, usr_fig3)
```
