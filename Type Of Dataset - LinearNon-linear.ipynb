+ <b> To understand the dataset is linear or non-linear we can have a experiment with linear regression. If the r2_score is coming good & the error (mae,mse,rmse) is less then the data is linear. Else the dataset is nonlinear & we have to chose non-linear / Polynomial regression algorithms to train the model

```python
# General imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generating data
X = np.random.randn(100,1)
c = np.random.uniform(-10,10,(100,))
# adding another non-linear column
X = np.hstack((X, X*X))

Y = (4*X[:,1] + c)

print('training data shape')
print(X.shape)

print('output training data shape')
print(Y.shape)

plt.scatter(X[:, 0], Y)
plt.show()
plt.scatter(X[:, 1], Y)
plt.show()
# Applying linear reg
from sklearn.linear_model import LinearRegression
regressor = LinearRegression().fit(X, Y)
# Checking the accuracy
from sklearn.metrics import r2_score
print(r2_score(regressor.predict(X), Y))
```
